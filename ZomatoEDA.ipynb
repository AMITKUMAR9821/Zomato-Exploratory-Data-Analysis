### Findind the top 10 cuisine types 

```python
cuisine_counts = final_df['Cuisines'].str.split(', ', expand=True).stack().value_counts()

top_cuisines = cuisine_counts.head(10)
plt.figure(figsize=(8, 4))
top_cuisines.plot(kind='bar', color='lightgreen')
plt.xlabel('Cuisine Type')
plt.ylabel('Number of Restaurants')
plt.title('Top 10 Popular Cuisine Types')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.show()
```

# Observation
Most of the customers likes north indian cuisine 

### Finding how do average ratings vary across different price ranges?

```python
plt.figure(figsize=(6, 4))
sns.boxplot(x='Price range', y='Aggregate rating', data=final_df, palette='pastel')
plt.xlabel('Price Range')
plt.ylabel('Rating')
plt.title('Rating Distribution Across Price Ranges')
plt.grid(True)
plt.show()
```

# Observation 

Lower-priced restaurants show a wider range of ratings (2 to 4.5), while higher-priced ones tend to have more consistent ratings (3.8 to 4.5), suggesting a positive correlation between price range and average ratings. The most expensive restaurants typically have the highest median ratings, often above 4.0.
